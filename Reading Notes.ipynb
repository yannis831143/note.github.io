## 机械复制时代的艺术作品

“将一系列的传统观念搁置一旁——比如创作与天赋，永恒价值与神秘等——因这些观念的引用并未受到控制（且目前尚难以控制），从而使得事物状况顺着法西斯的方向发展起来。”

“艺术作品以机械手法来复制却是个新近的现象，诞生于历史演进中，逐步发展而来，每一步的发展都有一段长久的间隔，而跳接的节奏是愈来愈快。”

“利用轮转的照相装置可以在摄影棚内把快如演员念台词的动作变化一一定在影像中。如果说石版潜藏孕育了画报，摄影则潜藏孕育了有声电影，电影是在摄影中萌芽成长的。”

“如同水、瓦斯和电流可从远方通到我们的住处，使我们毫不费劲便满足了我们的需求，有一天我们也将会如此得到声音影像的供应，只消一个信号、一个小小的手势，就可以让音像来去生灭。”

“在机械复制时代，艺术作品被触及的，就是它的“灵光”；这类转变过程具有征候性，意义则不限于艺术领域。也许可说，一般而论，复制技术使得复制物脱离了传统的领域。这些技术借着样品的多量化，使得大量的现象取代了每一事件仅此一回的现象。复制技术使得复制物可以在任何情况下都成为视与听的对象，因而赋予了复制品一种现时性。这两项过程对递转之真实造成重大冲击，亦即对传统造成冲击，而相对于传统的正是目前人类所经历的危机以及当前的变革状态。这些过程又与现今发生的群众运动息息相关。最有效的原动力就是电影。但是即使以最正面的形式来考量，且确实是以此在考量，如果忽略了电影的毁灭性与导泻性，也就是忽略了文化遗产中传统元素之清除的话，也就无法掌握电影的社会意义了。”

“一是有关作品的崇拜仪式价值，另一是有关其展览价值。艺术的生产最早是为了以图像来供应祭典仪式的需要。我们得承认这些图像的“在场”比它们被看见与否更为重要。石器时代的人在石窟墙上画的鹿是一种魔法媒介。这些壁画自然可以让别人看见，但主要还是为了神灵的注目而画的。后来，正是因诸如此类的祭仪价值而将艺术品藏匿于隐秘处；有些神祗的塑像是藏在教士才可进入的祈祷室之内。有些圣母像几乎是一年到头都被覆盖着不露面，也有些哥特式大主教堂的雕像从地面上根本望不见。”

# 中国古代哲学史大纲 胡适

“凡研究人生切要的问题，从根本上着想，要寻一个根本的解决，这种学问，叫做哲学。”

“一、天地万物怎样来的。（宇宙论）二、知识、思想的范围、作用及方法。（名学及知识论）三、人生在世应该如何行为。（人生哲学，旧称“伦理学”）四、怎样才可使人有知识，能思想，行善去恶呢。（教育哲学）五、社会国家应该如何组织，如何管理。（政治哲学）六、人生究竟有何归宿。（宗教哲学）”

“汉学的方法，只是用古训、古音、古本等等客观的根据，来求经典的原意。”

“述学是用正确的手段，科学的方法，精密的心思，从所有的史料里面，求出各位哲学家的一生行事、思想渊源沿革和学说的真面目。”

“思想 凡能著书立说成一家言的人，他的思想学说，总有一个系统可寻，决不致有大相矛盾冲突之处。”

“做历史的人，千万不可存一毫主观的成见。须知东西的学术思想的互相印证，互相发明，至多不过可以见得人类的官能心理大概相同，故遇着大同小异的境地时势，便会产出大同小异的思想学派。东家所有，西家所无，只因为时势境地不同，西家未必不如东家，东家也不配夸炫于西家。何况东西所同有，谁也不配夸张自豪。”

“第一步须搜集史料。第二步须审定史料的真假。第三步须把一切不可信的史料全行除去不用。第四步须把可靠的史料仔细整理一番：先把本子校勘完好，次把字句解释明白，最后又把各家的书贯串领会，使一家一家的学说，都成有条理有统系的哲学。做到这个地位，方才做到“述学”两个字。然后还须把各家的学说，笼统研究一番，依时代的先后，看他们传授的渊源，交互的影响，变迁的次序，这便叫做“明变”。然后研究各家学派兴废沿革变迁的原故，这便叫做“求因”。然后用完全中立的眼光，历史的观念，一一寻求各家学说的效果影响，再用这种种影响效果来批评各家学说的价值，这便叫做“评判”。”

“老子的最大功劳，在于超出天地万物之外，别假设一个“道”。这个道的性质，是无声、无形；有单独不变的存在，又周行天地万物之中；生于天地万物之先，又却是天地万物的本源。这个道的作用，是大道汜兮，其可左右。万物恃之而生而不辞，功成不名有，衣养万物而不为主。”

“老子把道与无看作一物，故他的哲学都受这种观念的影响”

“老子对于社会政治有两种学说：一是毁坏一切文物制度；一是主张极端放任无为的政策。”

“知识愈高，欲望愈难满足，又眼见许多不合意的事，心生无限烦恼，倒不如无知的草木，无思虑的初民，反可以混混沌沌，自寻乐趣。”

“上善若水，水利万物而不争。处众人之所恶，故几于道。”

“三个观念（一）易，（二）象，（三）辞，便是《易经》的精华。孔子研究那时的卜筮之《易》，竟能找出这三个重要的观念：第一，万物的变动不穷，都是由简易的变作繁赜的。第二，人类社会的种种器物制度礼俗，都有一个极简易的原起，这个原起，便是“象”。人类的文明史，只是这些“法象”实现为制度文物的历史。第三，这种种“意象”变动作用时，有种种吉凶悔吝的趋向，都可用“辞”表示出来，使人动用都有仪法标准，使人明知利害，不敢为非。”

“《秋水》、《庚桑楚》、《寓言》三篇最多可靠的材料”

“庄子的学说，只是一个“出世主义”

“疑独便是永远单独存在。”

“有一种“无”：无形、无色、无声、无味，却又是形声色味的原因；不生、不化，却又能生生化化。因为他自己不生，所以永久是单独的（疑独）。因为他自己不化，所以化来化去终归不变（往复）。这个“无”可不是老子的“无”了。老子的“无”是虚空的空处。《列子》书的“无”，是一种不生、不化，无形色声味的原质。一切天地万物都是这个“无”“自生”、自化、自形、自色、自智、自力、自消、自息”的结果。”

“万物皆种也，以不同形相禅”，这十一个字竟是一篇“物种由来”。他说万物本来同是一类，后来才渐渐的变成各种“不同形”的物类。却又并不是一起首就同时变成了各种物类。这些物类都是一代一代的进化出来的，所以说“以不同形相禅”。”

“自动的适合，是本来不适于所处的境遇，全由自己努力变化，战胜天然的境遇。如人类羽毛不如飞鸟，爪牙不如猛兽，鳞甲不如鱼鳖，却能造出种种器物制度，以求生存，便是自动的适合最明显的一例。《庄子》的进化论只认得被动的适合，却不去理会那更重要的自动的适合。”

“庄子因为不信辩论可以定是非，所以他的名学的第一步只是破坏的怀疑主义。”

“天下的是非，本来不是永远不变的。世上无不变之事物，也无不变之是非。”

“彼出于是，是亦因彼。……是亦彼也，彼亦是也。……彼亦一是非，此亦一是非。……是亦一无穷，非亦一无穷”。”

“以上所说，意在指点出庄子名学的一段真理。但是庄子自己把这学说推到极端，便生出不良的效果。他以为是非既由于偏见，我们又如何能知自己所见不偏呢？”

“无论什么都有存在的道理，既然如此，世上种种的区别，纵横、善恶、美丑、分合、成毁、……都是无用的区别了。既然一切区别都归无用，又何必要改良呢？又何必要维新革命呢？庄子因为能“达观”一切，所以不反对固有社会”

“既为世俗所通用，自然与世俗相投相得”

“庄子对于人生一切寿夭、生死、祸福，也一概达观，一概归到命定。”

“以为人力全无助进的效能，因此庄子虽说天道进化，却实在是守旧党的祖师。他的学说实在是社会进步和学术进步的大阻力。”

“孟子的政治学说含有乐利主义的意味”

“孟子所攻击的“利”字只是自私自利的利。大概当时的君主官吏都是营私谋利的居多。这种为利主义，与利民主义绝相反对。”

另外,以非破坏、非杀伤为前提的宙域的电磁脉冲攻击构成了威胁。

2030年代,中国因宙域战略落后于美国,宙域军和共产党的摩擦成为后来政变的契机。

Q.请告诉我2058年的未解决案件。

在数字日元和旧纸币并行使用的2031年,发生了战后最大规模(65亿日元)的假币流通事件。

是说旧纸币的稀有价值会提高,然后换成数字日元的手法。

特别是以珍贵旧纸币的年轻人为目标。

Q.日本的航空行业是怎样的?

虽然也有政府主导的行业重组,但面孔不变。

但是由于政治原因,面向台湾、华国的航空航班由大型公司各自运营。

另外,美国航空公司正在开通日本的一部分国内专用路线。

Q.请告诉我今后在日本的大新闻。

在福岛三陆近海发现了巨大的海底油田。据说如果再找到50年的话,世界和日本的关系就会颠覆。

Q.今后在日本大麻会合法化吗?

奢侈品在政府的严格管理下,从2040年代开始逐步合法化。

这是政府(大麻以外的)非法药物市场的完全根除和税收化为目的。

顺便说一下,只能从财政部管辖的特殊法人购买。感觉就像以前的专卖公社一样。

Q.SDGs到2030年能实现吗?

不幸的是,情况并非可以实现。特别是气候变化和部分国家的人权意识更加恶化了。

之后,以达到2045年为目标,日内瓦规定了可持续发展义务(Sustainable Development Obligations)。

Q.要准备什么移民到太空呢?

2024年,美国企业家将虚拟货币从地球向火星的探测器进行大量汇款。

即使在地球上很富有,金块和不动产也不能带到宇宙空间。

在宇宙中确立人类财富的保存方法,是宇宙迁移的关键。

Q.为了应对未来的灾害,能做什么?

2058年一般为“复合灾害”做准备。

如果台风+巨大地震和传染病+大规模停电同时袭来的话...?

也许还是讨论一下比较好。。

如果你认为这是未来人的警告。

Q.把这个保存在未来吧!有类似的东西吗?

2020年4月政府向全户分发的新品未开封的布制口罩,在这个时代的网上拍卖中以10万日元中标成为话题。

有“当时其实发了2次10万日元”的逸话。

Q.仅次于YouTube或TikTok,流行的视频服务是什么?

是bilibili。。根据美国的意向,广告限制变强,视频内容被限制的YouTube替换,以α世代为中心,用户数量将爆炸性地增加。

顺便说一下,现代不支持日语。

Q.未来的厕所和现在的厕所有什么区别吗?

家用的话可以通过排泄物检测健康状态,可以自动通知主治医生。

顺便说一下,在2058年,涩谷和新宿设置了非法药物检测型公共卫生间,以东京都的预算设置了。

Q.新冠肺炎这种传染病今后也会发生吗??

以猪为中间宿主的SARS-CoV-3(COVID-32)在2030年代中期被确认。然而,我们处理得比较早。

日本也作为新冠病毒的教训进行了省厅重组,进行传染病对策的垂直整合,所以损害被控制在最小范围内。

Q.今后基本收入会实现吗?

2027年只用数字日元可以实现小额。福利与我的号码绑定,每半年不使用的部分是自动返还的机制。

但是,由于不是对物价上涨增加补助金的机制,所以其有效性正在讨论中。

Q.希望能告诉我今后发生的社会问题。

由于只有网络新闻记录的文章的链接中断,丢失了在纸质媒体上没有记录的2020年代事件记录的现象成为社会问题。

这个问题也被称为“人类的记忆丧失”,后来也成为了纸质媒体重新评价的契机。

Q.未来数字标志会普及吗?

正在普及。只需将自己的OS设备放在口袋里,就可以在街道上自由操作画面。

但是,不能在街道上放大私密内容,相当于智能手机的设备也会在2058年留下。

Q.请告诉我2058年印度和中国的关系。

华国和印度的关系在经济上和军事上都变得极为恶劣。

虽然实质上实现了民主化,但是中华经济圈的世界霸权成为了华国的精神支柱。

特别是马六甲海峡围绕着宇宙电梯的权益而对立。

Q.在日本有问题吗?

不是“空房子”问题的“空空太阳能”问题变得严重。

特别是地方城市没有进行停止事业活动的太阳能发电设施的解体工作,在那个土地上作为垃圾滞留的情况不会断绝。

Q.工作方式有变化吗?

可以进行育儿休学(结婚、育儿休学的求职者也作为应届毕业生)制度。特别是女性是大学生,短期大学生期间结婚、生孩子后第一次就业的例子也增加了。

然而,在创立这个制度之前,给人一种相当大的矛盾的印象。

有了这个制度之后,女性为了分娩和育儿而不辞职,可以一直在同一家公司工作到退休年龄。顺带一提,引进这个制度的是日本第一位女性总理大臣。

Q.在2058年的求职活动中,有这个时代没有的职业吗?

一般职,所谓的办公岗位几乎没有了。这不是负面的意思,因为少子化导致劳动力不足严重,业务自动化扩大。

Q.以后假视频泛滥,年轻人离开视频是真的吗?

好像误会了,对不起。

严格来说,比起录影视频,无法篡改的LIVE视频的价值今后会飞跃地上升。

这个时代的YouTube已经变成那样了吧?

Q.请告诉我全球变暖的影响。

种植庄稼的环境在变化。简单来说,在西日本可以种植大米的地区相当有限。

Q.日本成为联合国的常任理事国了吗?

从终战开始看,在2045年里程碑的一年成为了常任理事国。

在此之前,中国体制的变化和北极海航路开通导致邻国地缘政治学的影响和理由多种多样。

Q.相反,现在有想告诉老年人的事情吗?

希望您能理解10多岁经历的1年的自肃和活到70岁的人经历的1年的重量和绝望感。

新冠疫情后的日本,这个故事被广泛讨论。他们也被称为“青春罗斯杰内世代”。

Q.听说到2050年为止可以进行“宇宙电梯”,是真的吗?

虽然还没有完成,但是赤道附近的新加坡有建设计划。日本的通用也有关系。

Q.来这边吓过一跳吗?

教科书的历史解释和实际的景象有很大的不同。

Q.我现在想成为YouTuber,但犹豫要不要像父母说的那样在大企业就业。

Q.2058年日本消费税是百分之几?

消费税是18%。但是生活必需品是15%。

Q.请告诉我今后美国的行动。

美军随时从以石油为资源的伊朗和阿富汗等中东撤退,以中国对太平洋的活动为主轴。

认为世界发达国家在脱石油、脱碳运动是那个准备。

Q.希望能告诉我今后的日本和中国、俄罗斯的关系。

中国在俄罗斯全贯通北极海航线后,对日本的态度软化。因为中国船不经过苏伊士运河,最短到达欧洲的话,需要经过日本的津轻海峡。

日俄关系也会通过对中战略戏剧性地改善。

Q.有想告诉现在的年轻人(10多岁)的事情吗?

我这一代都很年轻,虽然不是很了不起的立场,但是越是年轻人,为了防止感染扩散,还是好好自律比较好。

虽然父母正好是那一代,但那一代大家都后悔了。

Q.我想了解一下未来的LGBT和同性婚姻。

虽然在日本还没有承认同性婚姻,但LGBT在社会上成为了相当广泛认可的价值观。

2030年成立了GSM厅。这是从内阁府当时的LGBT措施推进会议开始发展的。

Q.未来北方领土问题会解决吗?

没有解决。俄罗斯推进千岛、北方领土的开发,开通了比苏伊士运河最短的欧洲北极航线。这是气候变化的副产品。

日本虽然事实上放弃了北方领土的归还,但是推进了苫小牧的开发,取得了很大的利益。

Q.外星人和外星生命体的存在到2058年才能确认吗?

在太阳系确认。好像是在木星或土星的卫星上发现的。不好意思,这是个不平的回答。但是未来不远啊。

Q.未来的银行怎么样了?

银行和现代一样没有实体店。引进数字日元后,全部在网上完成。

地银大多在当时的日银和金融厅的帮助下,勉强维持着。也有谷歌等外资进入,没有以前大银行的势头。

Q.2058年哪个国家有经济实力?

如果只看GDP的话,以微弱的差距,华国(现在的中国)是首位。第2号和美国、印度、德国、日本并列。如果只看成长率的话,印度是第一。

Q.希望能告诉我关于未来的便利店行业。

虽然店铺的面孔和现在没什么区别,但是除了城市以外的店铺几乎都是无人化的。

因为有用手机下单送到家里的系统,所以使用那个的人占压倒性多。

Q.这个国家的前途大致就好,请告诉我。

GDP是世界第5位(超过德国和印度),人口是8600万人。发生重大事件是汽车产业的衰退,反之的是脱碳对策和农林水产的高科技化,从世界开始一目了然。

Q.2058年的比特币价格是多少?

如果我的记忆正确的话,1BTC=约7亿8000万日元。在我这个时代,交易只有卖出,不能新购买。

Q.有新国家的出现吗?

欧洲的话是加泰罗尼亚和苏格兰,亚洲是台湾。

Q.飞翔的车是实用化的吗?

飞翔的车已经实用化了。但是,其使用仅限于紧急车辆和地方和离岛的物流。

Q.汽车行业怎么样了?

所有这些都变成了电动汽车。2058年国产车只有丰田和本田。

日产虽然只留在欧洲作为品牌,但马自达、速霸陆和三菱汽车在电动化的浪潮中没能生存下来。

Q.今后会有富士山喷发和南海海沟地震吗?

没有。记录有2024年1月的“令和6年十胜冲地震”的广域地震和2037年1月的“令和19年冲绳地震”的城市直下型地震。

特别是后者是同年8月的余震发生和台风袭击同时袭击的悲剧性,也是重新审视建筑标准法的机会。

Q.未来的iPhone怎么样了?

2058年也有相当于iPhone系列的设备。厚度是2mm,像信用卡一样的尺寸。

和现在不一样,这个时代不需要钱包,所以放在卡套里使用比较流行。

Q.在你生活的世界里,中国怎么样了?

在我们这个时代,为了和现在的中国政府区分开来,被称为华国。由于2030年代后期的国内纷争,长期处于多重政府状态。也有人民解放军和别的华国军,这话有点长,后述。

在这样的背景下,国内萧条一直持续到2030年中期。共产党也因为不熟练地加强对企业的经济干预而引发了企业家的反对。

之后,中华企业开始背后煽动民主化运动,部分激进派黑客入侵军队系统。也有人说空军的创新军官参与了。

2024年9月宣布结束。
